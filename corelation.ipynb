```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fortnite = pd.read_csv('Fortnite Statistics.csv', index_col=0)
```

```python
fortnite.head()
```

```python
correlation = fortnite[['Eliminations', 'Assists', 'Revives', 'Hits', 'Damage to Structures']].corr(method='pearson', numeric_only=False)
correlation.style.background_gradient(cmap='coolwarm').format(precision=2)
sns.heatmap(correlation)
plt.show()
```
